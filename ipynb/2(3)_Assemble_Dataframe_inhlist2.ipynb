```python
# Use path to create variables describing experimental conditions

vars=['inh','inj','exp','date','mouse']
#pc.insert(len(pc.columns),n,0) for n in vars
for n in vars:
    pc.insert(len(pc.columns),n,'NaN')

fpos_inh=3
fpos_inj=2
#fpos_exp=4
fpos_date=0
fpos_mouse=1
f_end=4

for p in pd.unique(pc.path):
    n=Path(p).stem
    #print(str(n.stem).split('_')[fpos_inh])
    pc.loc[(pc.path==p),'inh']=str(n.split('_')[fpos_inh])
    pc.loc[(pc.path==p),'inj']=int(n.split('_')[fpos_inj][3:])
    #pc.loc[(pc.path==p),'exp']=int(n.split('_')[fpos_exp][3:4])
    pc.loc[(pc.path==p),'date']=str(n.split('_')[fpos_date][:])
    pc.loc[(pc.path==p),'mouse']=int(n.split('_')[fpos_mouse][5:])
    pc.loc[(pc.path==p),'path']=str(n)

#pc['mouse']=pc.mouse.astype('int')

# df.drop(columns=['B', 'C'])
#columns=['B', 'C'])
```
